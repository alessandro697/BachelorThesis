```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fft import fft, fftfreq
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.transforms as mtransforms
```

```python

my_file = pd.read_csv("output___n_2__gam_10.0__U0_1.0__F_7.5__R_0.25__R2_0.14433756__R3_0.25__c_1.0.dat"
                      ,sep=' ')
data = np.array(my_file) 
R1 = np.sqrt(3)/12
R2 = 0.25
c = 1
r = data[:,3]-data[:,2]
print(len(r))
xcm = (data[:,3]+data[:,2])/2
start=20000
end=300000
#forza sin(r)cos(xcm)
y = -2*np.pi*np.sin(np.pi*(-data[start:end,2]+data[start:end,3]))*np.cos(np.pi*(data[start:end,2]+data[start:end,3]))

vcm1 = np.mean(data[start:end,4]+data[start:end,5])/2
#print('La velocità del centro di massa è : ' , vcm)

w = pow(12*c*(pow(R2,4) - pow(R2,2)*pow(R1,2)), 1/2)
print('La frequenza armonica in un intorno di R2 è : ' ,w)

R1 = fft(r[start:end])
freqr1 = fftfreq(end-start)*100
plt.plot(freqr1, np.abs(R1),label='frequency')
plt.plot([vcm1,vcm1],[0,50],label='Vcm',linestyle='dashed')

plt.xlim([0.4,2.5])
plt.ylim([0,980])
plt.show()

Y1 = fft(y)
freqr11 = fftfreq(len(y))*100
plt.plot(freqr11, np.abs(Y1),label='frequency')
#plt.plot([vcm,vcm],[0,1600],label='Vcm',linestyle='dashed')
plt.xlim([0.4,2.5])
plt.ylim([0,44900])
plt.show()
```
