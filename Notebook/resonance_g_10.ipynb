```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fft import fft, fftfreq
import matplotlib.transforms as mtransforms
```

```python
R1 = 1.5
R2 = 2.3
c = 0.001

w = pow(12*c*(pow(R2,4) - pow(R2,2)*pow(R1,2))*2, 1/2)/(2*np.pi)
print('La frequenza armonica in un intorno di R2 è : ' ,w)

delta = (c/2.)*(pow(R2,6)-pow(R1,6)+3*pow(R2,2)*pow(R1,4)-3*pow(R2,4)*pow(R1,2))
print('La barriera di potenziale è : ' ,delta)

start=2500
end=3000

Fmin = 2
Fmax = 3
df = 0.1
Forces1 = np.linspace(Fmin, Fmax , int((Fmax-Fmin)/df +1))

Fmin = 3
Fmax = 3.5
df = 0.01
Forces2 = np.linspace(Fmin, Fmax , int((Fmax-Fmin)/df +1))

Fmin = 3.5
Fmax = 10
df = 0.1
Forces3 = np.linspace(Fmin, Fmax , int((Fmax-Fmin)/df +1))

Forces = np.concatenate((Forces1,Forces2, Forces3))
Vel_cm = np.zeros(len(Forces))
Ampiezze = np.zeros(len(Forces))
#print(Forces)

i=0
for F in Forces:
    my_file = pd.read_csv("output___n_2__gam_10.0__U0_1.0__F_"+str(F)+"__R_2.3__R2_1.5__R3_2.3__c_0.001.dat"
                      ,sep=' ')
    data = np.array(my_file)
    
    r = -data[:,2]+data[:,3]
    #calcolo ampiezza oscillazione
    rmax = np.max(r[start:end])
    rmin = np.min(r[start:end])
    A = rmax - rmin
    Ampiezze[i] = A
    
    vcm = (data[end-1,2]+data[end-1,3]-data[start,2]-data[start,3])/(2*(data[end-1,0]-data[start,0]))
    Vel_cm[i] = vcm
    
    i+=1
    

plt.xlabel('Vcm')
plt.ylabel('Amplitude')
plt.title('Resonance: g=10 , c=0.001')
plt.plot(Forces, Ampiezze, linewidth=1, marker ='.')
#plt.plot([w,w], [0,0.5], label = 'w', linestyle = 'dashed')
plt.grid()
#plt.xlim([3.1,3.5])
#plt.ylim([0.0,0.13])
plt.show()

Max = Vel_cm[np.where(Ampiezze == np.max(Ampiezze))[0][0]]
print('Il picco di ampiezza vale:', np.max(Ampiezze))
print('Il rapporto tra la frequenza del picco e quella armonica è :' , Max/w)

Forces_0_001 = Forces
Vel_cm_0_001 = Vel_cm
Ampiezze_0_001 = Ampiezze
w_0_001 = w
```
