```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
R1 = np.sqrt(3)/4
R2 = 0.75
R3 = 0.7
R4 = 0.8
c = 1

x = np.linspace(0,R2+0.5, 10000)
delta = (c/2.)*(pow(R2,6)-pow(R1,6)+3*pow(R2,2)*pow(R1,4)-3*pow(R2,4)*pow(R1,2))
V1 = c*(x**6 - 1.5*(R1**2 + R2**2)*x**4 + 3*(R1*R2*x)**2 )
V2 = c*(x**6 - 1.5*(R1**2 + R3**2)*x**4 + 3*(R1*R3*x)**2 )
V3 = c*(x**6 - 1.5*(R1**2 + R4**2)*x**4 + 3*(R1*R4*x)**2 )


fig, ax = plt.subplots(1)
ax.set(xlabel='$r \quad [a]$')
ax.set(ylabel='$V_{int} \quad [V_0] $')


ax.plot(x, V1 , label='$R_2 = 0.75 a$',linewidth=1, color='black')
ax.plot(x, V2 , label='$R_2 = 0.7 a$',linewidth=1, color='orangered', linestyle='dashed')
ax.plot(x, V3 , label='$R_2 = 0.8 a$',linewidth=1, color='darkblue', linestyle='dashdot')
ax.plot([R1,R1],[-4,5],linestyle='dashed', color='forestgreen',linewidth=1)
ax.set(xlim=[0,R4+0.2])
ax.set(ylim=[-0.018,0.06])
ax.text(0.38, 0.05, '$R_1$',color='forestgreen')

axR = ax.twinx()
axT = ax.twiny()
ax.tick_params(direction='in')
axT.tick_params(direction='in')
axR.tick_params(direction='in')
axR.yaxis.set_major_formatter(plt.NullFormatter())
axT.xaxis.set_major_formatter(plt.NullFormatter())
axR.set(ylim = [-0.018,0.06])
axT.set(xlim=[0,R4+0.2])
axR.set(xlim=[0,R4+0.2])
ax.legend()

plt.savefig("V_int.pdf", dpi=300, bbox_inches='tight')
plt.savefig("V_int.png", dpi=300, bbox_inches='tight')
plt.show()
```
