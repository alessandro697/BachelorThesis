```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib.transforms as mtransforms

my_file = pd.read_csv("output___n_2__gam_10.0__U0_1.0__F_5.0__R_0.75__R2_0.4330127__R3_0.75__c_1.0.dat"
                      ,sep=' ')
data = np.array(my_file) 

R1 = np.sqrt(3)/4
R2 = 0.75
c = 1
r = data[:,3]-data[:,2]
xcm = (data[:,3]+data[:,2])/2
delta = (c/2.)*(pow(R2,6)-pow(R1,6)+3*pow(R2,2)*pow(R1,4)-3*pow(R2,4)*pow(R1,2))
print('La barriera di potenziale Ã¨ : ' ,delta)

plt.xlabel('t')
plt.ylabel('x1-x2')
plt.title('Fig 11 (b)')
plt.plot(data[:,0], r , label='x2-x1',linewidth=0.2)
plt.grid()
plt.plot(data[:,0], 0.75*np.ones(len(data[:,0])), label='R2',linestyle='dashed')
plt.plot(data[:,0], (np.sqrt(3)/4)*np.ones(len(data[:,0])), label='R1',linestyle='dashed')
plt.legend()
plt.show()



#xcm e r sullo stesso grafico
fig, (ax1, ax2) = plt.subplots(2, sharex=True)

ax1.plot(data[:,0], r , label='$x_2-x_1$', color ='darkblue',linewidth=1)
ax1.plot(data[:,0], R2*np.ones(len(data[:,0])),'orangered', label='$R_2$',linestyle='dashed')
ax1.plot(data[:,0], R1*np.ones(len(data[:,0])), 'forestgreen',label='$R_1$',linestyle='dashed')
ax1.text(46, 0.82, '$R_2$',color='orangered')
ax1.text(46, 0.34, '$R_1$', color='forestgreen')
ax1.set(ylabel='$x_2-x_1 \quad [a]$', ylim = [0.31,1])

ax1R = ax1.twinx()
ax1T = ax1.twiny()
ax1.tick_params(direction='in')
ax1T.tick_params(direction='in')
ax1R.tick_params(direction='in')
ax1R.yaxis.set_major_formatter(plt.NullFormatter())
ax1T.xaxis.set_major_formatter(plt.NullFormatter())
ax1R.set(ylim = [0.31,1])
ax1R.set(xlim=[0,100])


ax2.plot(data[:,0], xcm , label='$x_2-x_1$', color ='darkblue',linewidth=1)
ax2.set(ylabel='$x_{cm} \quad [a]$', ylim = [0,48])

ax2R = ax2.twinx()
ax2T = ax2.twiny()
ax2.tick_params(direction='in')
ax2T.tick_params(direction='in')
ax2R.tick_params(direction='in')
ax2R.yaxis.set_major_formatter(plt.NullFormatter())
ax2T.xaxis.set_major_formatter(plt.NullFormatter())
ax2R.set(ylim = [0,48])
ax2R.set(xlim=[0,100])

ax2.set(xlabel='$t \quad [ a\sqrt{m/V_0} ] $')

trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
ax1.text(0.0, 1.0, '(a)', transform=ax1.transAxes + trans,
            fontsize='medium', verticalalignment='top', fontfamily='serif',
            bbox=dict(facecolor='0.8', edgecolor='none', pad=3.0))
ax2.text(0.0, 1.0, '(b)', transform=ax2.transAxes + trans,
            fontsize='medium', verticalalignment='top', fontfamily='serif',
            bbox=dict(facecolor='0.8', edgecolor='none', pad=3.0))

plt.subplots_adjust(wspace=0, hspace=0)
plt.xlim([0,100])
plt.savefig("11_b_R_Xcm.pdf", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import scipy.integrate

start = 4500
end = 5000
#calcolo energia immessa
y = -2*np.pi*np.sin(np.pi*(-data[start:end,2]+data[start:end,3]))*np.cos(np.pi*(data[start:end,2]+data[start:end,3]))
P = y*(-data[start:end,4]+data[start:end,5])
energy = scipy.integrate.trapz(P, data[start:end,0])
print(energy)
spaces = np.arange(100)
e1 = np.zeros(100)

for i in spaces:
    term = int(start + (i+1)*(end-start)/len(spaces))
    e1[i] = scipy.integrate.trapz(P[:int((i+1)*(end-start)/len(spaces))], data[start:term,0])
    
term = start + (spaces+1)*(end-start)/len(spaces)
plt.xlabel('T')
plt.ylabel('W')
plt.title('Exact work')
plt.plot(term,e1,linewidth=1)
plt.grid()
plt.show()
```

```python
#ora con approssimazione cos(wt)
y2 = -2*np.pi*np.sin(np.pi*(-data[start:end,2]+data[start:end,3]))*np.cos(np.pi*(data[start,2]+data[start,3]+2*vcm*data[0:end-start,0]))
vcm = np.mean(data[start:end,4]+data[start:end,5])/2
print(vcm)



P2 = y2*(-data[start:end,4]+data[start:end,5])
energy2 = scipy.integrate.trapz(P2, data[start:end,0])
print(energy2)
e2 = np.zeros(100)

for i in spaces:
    term = int(start + (i+1)*(end-start)/len(spaces))
    e2[i] = scipy.integrate.trapz(P2[:int((i+1)*(end-start)/len(spaces))], data[start:term,0])
    
term = start + (spaces+1)*(end-start)/len(spaces)
plt.xlabel('T')
plt.ylabel('W')
plt.title('Approximate work')
plt.plot(term,e2,linewidth=1)
plt.grid()
plt.show()


plt.xlabel('T')
plt.ylabel('(W2-W1)/(W2+W1)')
plt.title('Normalized Difference')
plt.plot(term,(-e2+e1)/(e2+e1),linewidth=1)
plt.grid()
plt.show()
```
