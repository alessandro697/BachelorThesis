```python
R1 = 1.5
R2 = 2.3
c = 0.05

w = pow(12*c*(pow(R2,4) - pow(R2,2)*pow(R1,2))*2, 1/2)/(2*np.pi)
print('La frequenza armonica in un intorno di R2  : ' ,w)

delta = (c/2.)*(pow(R2,6)-pow(R1,6)+3*pow(R2,2)*pow(R1,4)-3*pow(R2,4)*pow(R1,2))
print('La barriera di potenziale  : ' ,delta)

Fmin = 1.0
Fmax = 5.5
df = 0.1

start=2000
end=3000

Forces = np.linspace(Fmin, Fmax , int((Fmax-Fmin)/df +1))
Vel_cm = np.zeros(len(Forces))
Ampiezze = np.zeros(len(Forces))

i=0
for F in Forces:
    my_file = pd.read_csv("output___n_2__gam_5.0__U0_1.0__F_"+str(F)+"__R_2.3__R2_1.5__R3_2.3__c_0.05.dat"
                      ,sep=' ')
    data = np.array(my_file)
    
    r = -data[:,2]+data[:,3]
    #calcolo ampiezza oscillazione
    rmax = np.max(r[start:end])
    rmin = np.min(r[start:end])
    A = rmax - rmin
    Ampiezze[i] = A
    
    vcm = np.mean(data[start:end,4]+data[start:end,5])/2
    Vel_cm[i] = vcm
    
    i+=1
    

plt.xlabel('Vcm')
plt.ylabel('Amplitude')
plt.title('Resonance: g=5 , c=0.05')
plt.plot(Vel_cm, Ampiezze, linewidth=1, marker ='.')
plt.plot([w,w], [0,0.5], label = 'w', linestyle = 'dashed')
plt.grid()
#plt.xlim([0.2,1])
#plt.ylim([0.2,0.5])
plt.show()

Max = Vel_cm[np.where(Ampiezze == np.max(Ampiezze))[0][0]]
print('Il picco di ampiezza vale:', np.max(Ampiezze))
print('Il rapporto tra la frequenza del picco e quella armonica  :' , Max/w)

    
```
